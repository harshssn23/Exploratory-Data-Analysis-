```python
ab=['Age','Fare']
for i in ab:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1

    lb=q1-(1.5*iqr)
    ub=q3+(1.5*iqr)


    df= df[~((df[i]<=lb)|(df[i]>=ub))]
```

```python
for i in ab:
    sns.boxplot(data=df,x=i)
    plt.show()
```

```python
from statsmodels.stats.weightstats import ztest as ztest

print("H0: There is no diffrence betwen the means")
print("H1: There is a significant diffrence betwen the means")

l=df['Age']

a,b=ztest(l,value=28)

if b<0.05:
    print("Accept H1")
else:
    print("Accept H0")

```

```python
from statsmodels.stats.weightstats import ztest as ztest

print("H0: There is no diffrence betwen the means")
print("H1: There is a significant diffrence betwen the means")

l=df['Age']
m=df['Fare']

a,b=ztest(l,m)

if b<0.05:
    print("Accept H1")
else:
    print("Accept H0")

```

```python
from scipy import stats

print("H0: There is no diffrence betwen the means")
print("H1: There is a significant diffrence betwen the means")

l=df['Age']

a,b=stats.ttest_1samp(l,28)

if b<0.05:
    print("Accept H1")
else:
    print("Accept H0")

```

```python
from scipy import stats

print("H0: There is no diffrence betwen the means")
print("H1: There is a significant diffrence betwen the means")

l=df['Age']
m=df['Fare']

a,b=stats.ttest_ind(l,m)

if b<0.05:
    print("Accept H1")
else:
    print("Accept H0")

```

```python
from scipy.stats import f_oneway

print("H0: There is no diffrence betwen the means")
print("H1: There is a significant diffrence betwen the means")

l='Survived'
m=['Fare','Age']

for i in m:
    a=[df[i][df[l]==group]for group in df[l].unique()]
    
    a,b=f_oneway(*a)

    if b<0.05:
        print("Accept H1")
    else:
        print("Accept H0")

    print(b)

```
